Problem-1: Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
df=pd.read_csv('https://tinyurl.com/2fe6vz4u')
```

```python
df.head()
```

```python
Iran=df.query('Country=="Iran"').groupby('Year')['PM2.5'].sum()
China=df.query('Country=="China"').groupby('Year')['PM2.5'].sum()

plt.plot(Iran.index,Iran.values,label='Iran',linestyle='dashed')
plt.plot(China.index,China.values,label='China')
plt.title('PM2.5 OVER THE YEARS')
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.xticks(df['Year'].value_counts().index)
plt.legend()
plt.grid()
plt.show()

```

Problem-2: Draw a histogram of the column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'],density=True,facecolor='green',bins=50,alpha=0.6)
plt.xlabel('Bins')
plt.ylabel('Probability')
plt.title('Histogram of PM10')
plt.grid()
plt.show()
```

Problem-3: Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
Poland=df.query('Country=="Poland"')
Chile=df.query('Country=="Chile"')
plt.scatter(Poland['PM2.5'],Poland['PM10'],marker='+',label='Poland',color='black')
plt.scatter(Chile['PM2.5'],Chile['PM10'],label='Chile',color='red')
plt.title('PM2.5 Vs PM10 for Chile and Poland')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.legend()
plt.grid()
plt.show()
```

Problem-4: Draw a pie chart of top 5 most frequent countries.

```python
top5=df['Country'].value_counts().head(5)
plt.pie(top5,labels=top5.index,autopct='%0.1f%%')
plt.title('Top 5 most frequent countries')
plt.show()
```

Problem-5: Draw a bar chart which represents the counts of top 5 most frequent countries.

```python
plt.bar(top5.index,top5,label=top5.index,color=['green','brown','purple','black','red'])
plt.title('Top 5 countries bar chart')
plt.xlabel('Countries')
plt.ylabel('Frequency Count')
plt.legend()
plt.show()
```

Problem-6
Show a line plot of Total Profit for each month with below styling.

Dotted Line
Line Color Blue
Show Legend at top left
Circle Marker

```python
sales=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTJh6X4_mqixWsfK9mgkllGQkKYW9Wj9kOIMGY2uYsWeS8n5np87DO-SDGQWJ1HXEnxiOVFVzYFYEcR/pub?gid=558678488&single=true&output=csv')
```

```python
plt.plot(sales['month_number'],sales['total_profit'],label='Month on month profit',linestyle='dotted',color='blue',marker='o')
months=['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']
plt.xticks(sales['month_number'],months,rotation='vertical')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.title("Company sales profit")
plt.legend(loc='upper left')
plt.show()
```
