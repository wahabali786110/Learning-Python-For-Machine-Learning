```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snc
```

2-D Line Plot

```python
# plotting a simple function
price=[54000,76000,87000,43000,21000]
year=[2020,2021,2022,2023,2024]
plt.plot(year,price)
```

```python
# from a pandas dataframe 
# title/xlabel/ylabel/color=
runs=pd.read_csv('sharma-kohli.csv')
plt.plot(runs['index'],runs['V Kohli'],color='black')
plt.plot(runs['index'],runs['RG Sharma'],color='red')
plt.title('Rohit Sharma vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# linestyle=/dashed/dotted/dashdot/solid(default)
# linewidth=

# marker="D"/"+"/"."/"o"
# markersize=
runs=pd.read_csv('sharma-kohli.csv')
plt.plot(runs['index'],runs['V Kohli'],color='black',linestyle='solid',linewidth=5,marker='D',markersize=10)
plt.plot(runs['index'],runs['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='o',markersize=10)
plt.title('Rohit Sharma vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# legend -> location: can be given as an argument in the legend function
runs=pd.read_csv('sharma-kohli.csv')
plt.plot(runs['index'],runs['V Kohli'],color='black',linestyle='solid',linewidth=5,marker='D',markersize=10,label='Virat Kohli')
plt.plot(runs['index'],runs['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='o',markersize=10,label='Rohit Sharma')
plt.title('Rohit Sharma vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
```

```python
# limiting axes
price=[54000,76000,87000,43000,21000,4500000]
year=[2020,2021,2022,2023,2024,2025]
plt.plot(year,price)
plt.ylim(0,100000)
# plt.xlim(2021,2023)
```

Scatter Plots

```python
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)
```

```python
# plt.scatter using simple function
plt.scatter(x,y)
```

```python
# plt.scatter using pandas dataframe
# size: circles will be big if runs are high and smaller for less runs
batter=pd.read_csv('batter.csv')
batter=batter.head(50)
plt.scatter(batter['avg'],batter['strike_rate'],color='red',marker='o',s=batter['runs']/20)
plt.title('Top 50 batsman analysis with most runs')
plt.xlabel('Avg')
plt.ylabel('SR')
```

```python
tips=snc.load_dataset('tips')
# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# plotting usin plot 
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

Bar Chart

```python
# using dataframe
df=pd.read_csv('batsman_season_record.csv')
plt.bar(df['batsman'],df['2015'])
```

```python
# Way to plot multiple bar charts
df=pd.read_csv('batsman_season_record.csv')
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```
