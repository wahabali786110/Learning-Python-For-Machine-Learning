Q-1: Using groupby make groups using the "Pclass" column and find out the average age and total number of missing values in the "Age" column for every group

```python
pclass=titanic_df.groupby('Pclass')

pclass.agg({
    'Age':['mean',lambda x:x.isnull().sum()]
})

pclass.agg(
    Average_Age=('Age','mean'),
    Missing_Ages=('Age',lambda x:x.isnull().sum())
)


```

Q-2: Using groupby make groups using the "Pclass" column and fill every group's "Embarked" column's missing values with the mode value of that group. After that, print every group's "Embarked" column's value counts in ascending order.

```python
titanic_df['Embarked']=pclass['Embarked'].apply(lambda x:x.fillna(x.mode()[0])).reset_index(level=0,drop=True)
pclass['Embarked'].value_counts().sort_values()
```

Q-4: Perform groupby based on the "Team" column and then perform Z Normalization on top of the below columns of each group:
Passes
Passes Completed
Attempted Line Breaks
Completed Line Breaks
You have to make a python function named z_normalization which takes two arguments:

group: Every group that you have created
cols_to_perform: This parameter takes a list of columns on which you have to perform the Z-Normalization.
For this problem, you have to use th apply() method.

Z − Normalization=Xi−μstd 

After that find out the below values for each group:

minimum "Passess"
maximum "Passess"
minimum "Yellow Cards"
maximum "Yellow cards"
average "Yellow Cards"
maximum "Attempted Line Breaks"
minimum "Attempted Line Breaks"
standardard deviation of "Attempted Line Breaks"
average Possession

```python
def z_normalization(group,column_list):
    mean_val=group[column_list].mean()
    std=group[column_list].std()

    group[column_list]=(group[column_list]-mean_val)/std
    return group

column_list=['Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks']
teams=fifa_df.groupby('Team')

normalized_df=teams.apply(z_normalization,column_list=column_list)
normalized_teams=normalized_df.groupby('Team')

normalized_teams.agg({
    'Passes':['min','max'],
    'Yellow Cards':['min','max','mean'],
    'Attempted Line Breaks':['max','min','std'],
    'Possession (%)':'mean'
})
```

Q-5: Find batsman in below category-
1- Highest score while chasing

```python

chassing_df=ipl[ipl['inning']==2]
batsman_inning=chassing_df.groupby(['match_id','batsman'])
batsman_inning['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

2- Best Strike rate while chasing and have faced 100+ balls

```python
def best_strike_rate(group):
    balls_faced=group['ball'].size

    if balls_faced<100:
        return np.nan
    
    total_score=group['batsman_runs'].sum()

    return (total_score/balls_faced)*100

batsman_group=chassing_df.groupby('batsman')
batsman_group.apply(best_strike_rate).dropna().sort_values(ascending=False).head(1)
```

Q-6 Most Successful bowler against any batsman. Find that pair of bowler and batsman.
Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

```python

bowler_wickets = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled']


dismissals_df = ipl[(ipl['player_dismissed'] == ipl['batsman']) & 
                    (ipl['dismissal_kind'].isin(bowler_wickets))]


true_wickets = dismissals_df.groupby(['bowler', 'batsman']).size().reset_index(name='true_dismissals')

runs_conceded = ipl.groupby(['bowler', 'batsman'])['batsman_runs'].sum().reset_index()


final_comparison = pd.merge(true_wickets, runs_conceded, on=['bowler', 'batsman'])


winner_pair = final_comparison.sort_values(
    by=['true_dismissals', 'batsman_runs'], 
    ascending=[False, True]
).head(1)

print(winner_pair)
```

Q-7: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.

```python
ipl[['Player_1','Player_2']]=np.sort(ipl[['batsman','non_striker']],axis=1)
ipl.groupby(['Player_1','Player_2'])['total_runs'].sum().reset_index().sort_values(by='total_runs',ascending=False).head(1)
```

Q-8: Make a dataframe for all batting pairs played together.
Batsman1 Batsman2 Runs Avg StrikeRate
Just to ease this question you can count wide-balls for strike rate.

```python
new_df=ipl.groupby(['Player_1','Player_2']).agg(
    Runs= ('total_runs','sum'),
    Avg= ('total_runs','mean'),
   
).reset_index()

new_df['StrikeRate']=new_df['Avg']*100
new_df.sort_values(by='Runs',ascending=False,inplace=True)
new_df.rename(columns={'Player_1':'Batsman1','Player_2':'Batsman2'},inplace=True)
new_df

```
