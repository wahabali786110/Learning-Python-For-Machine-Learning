Group By:
Used to group columns on the basis of category and numeric values

```python
#Applying builtin aggergate functions on groupby objects
genre=movies.groupby('Genre')
genre.sum(numeric_only=True)
```

```python
#Find the top 3 genres by total earning
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
#find the genre with highest avg imdb rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
#find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
# movies['Star1'].value_counts()
```

```python
# Group By Attributes and Method:
# find total number of groups -> Length
len(genre)

# find items in each group -> Size
genre.size()

# first()/last()->nth item
genre.first()
genre.last()
genre.nth(6).set_index('Genre')
```

```python
#get_group -> vs filtering: to fetch a particulat group
genre.get_group('Animation')
```

```python
#Groups:
genre.groups
```

```python
# describe
# sample
# nunique
genre.describe()
genre.sample(2,replace=True)
genre.nunique()
```

```python
# agg method
#passing dict
genre.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# passing list
genre[['Runtime','IMDB_Rating']].agg(['mean','max','min'])
```

```python
#Adding both the syntax
genre.agg({
    'Runtime':['mean','sum'],
    'IMDB_Rating':['mean','sum'],
    'No_of_Votes':['sum','mean'],
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# find the highest rated movie in each genre
# looping on groups


highest_rated_movies=[]
for groups,data in genre:
    top_movie=data[data['IMDB_Rating']==data['IMDB_Rating'].max()]
    highest_rated_movies.append(top_movie)

df=pd.concat(highest_rated_movies)
df
```

```python
# split (apply) combine
# apply -> builtin function

# find the number of movies starting with A in each group

def movies_count(group):
    return group['Series_Title'].str.startswith('A').sum()

genre.apply(movies_count).sort_values(ascending=False)
```

```python
# find ranking of each movie in the group according to the IMDB rating

def ranking(group):
     group['Rank']=group['IMDB_Rating'].rank(ascending=False)
     return group

genre.apply(ranking).reset_index()
```

```python
# find normalized IMDB Ranking for each group
def normalized_rating(group):
    group['Normalized_Rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group

genre.apply(normalized_rating).reset_index()
```

```python
# groupby on multiple columns
duo=movies.groupby(['Director','Star1'])
duo.size().sort_values(ascending=False)
duo.get_group(('Akira Kurosawa','Toshirô Mifune'))
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best (in terms of metascore avg) actor-genre combo
actor_genre=movies.groupby(['Star1','Genre'])
actor_genre['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(7)
```
